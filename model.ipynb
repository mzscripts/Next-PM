# Election Prediction Features and Target Variables

## Features (Predictor Variables)
These are the columns that provide information about the election dynamics and will be used to make predictions:

- **Party**: Name of the party or alliance.
- **Alliance**: Indicates which alliance the party belongs to (e.g., NDA, INDIA alliance).
- **Popular vote Votes**: Total number of popular votes received.
- **Popular vote %**: Percentage of total popular votes received.
- **Vote Share Change (%)**: Change in vote share compared to the previous election.
- **Seats Contested**: Number of seats contested by the party.
- **Campaign Expenditure (INR in Crores)**: Total campaign expenditure.
- **Number of Rallies**: Total rallies conducted during the campaign.
- **Social Media Mention**: Number of social media mentions.
- **Sentiment Score**: Sentiment score of the public toward the party or candidate.
- **Key Performing States**: The primary regions of influence for the party.

## Target Variables
- **Next PM of India**: This is a categorical variable (e.g., Narendra Modi, Mamata Banerjee, etc.).
- **Winning Party**: This is also a categorical variable (e.g., Bharatiya Janata Party, Indian National Congress, etc.).



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
plt.style.use('seaborn-v0_8-darkgrid')

# Create a figure with multiple subplots
fig = plt.figure(figsize=(20, 15))

# 1. Popular Vote Distribution
plt.subplot(2, 2, 1)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(ds['Popular vote %'], labels=ds['Party'], autopct='%1.1f%%', colors=colors)
plt.title('Popular Vote Distribution (%)', pad=20)

# 2. Seats Won vs Contested
plt.subplot(2, 2, 2)
x = range(len(ds['Party']))
width = 0.35
plt.bar(x, ds['Seats Contested'], width, label='Seats Contested', color='#66b3ff')
plt.bar([i + width for i in x], ds['Seats Won'], width, label='Seats Won', color='#ff9999')
plt.xticks([i + width / 2 for i in x], ds['Party'], rotation=45)
plt.legend()
plt.title('Seats Contested vs Won')

# 3. Popular Vote Percentage by Party
plt.subplot(2, 2, 3)
sns.barplot(data=ds, x='Party', y='Popular vote %', palette='viridis')
plt.title('Popular Vote Percentage by Party')
plt.xlabel('Party')
plt.ylabel('Popular Vote (%)')
plt.xticks(rotation=45)

# 4. Vote Share Change (%) by Party
plt.subplot(2, 2, 4)
sns.barplot(data=ds, x='Party', y='Vote Share Change (%)', palette='viridis')
plt.title('Vote Share Change (%)')
plt.xlabel('Party')
plt.ylabel('Vote Share Change (%)')
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout(pad=3.0)

# Add a title to the entire figure
fig.suptitle('2024 Indian Election Analysis', fontsize=16, y=1.02)

# Display the plot
plt.show()

# Additional statistics
print("\nKey Statistics:")
print(f"Total votes analyzed: {ds['Popular vote Votes'].sum():,}")
print(f"Average campaign expenditure: ₹{ds['Campaign Expenditure (INR in Crores)'].mean():,.2f} Crores")
print(f"Party with highest sentiment score: {ds.loc[ds['Sentiment score'].idxmax(), 'Party']}")
print(f"Total seats contested: {ds['Seats Contested'].sum()}")
print(f"Total seats won: {ds['Seats Won'].sum()}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Define Features (X) and Targets (y1 and y2)
X = ds.drop(columns=['Party', 'Popular vote %', 'Seats Won'])  # Exclude target variables
y1 = ds['Popular vote %']  # Target for Vote Share (%)
y2 = ds['Seats Won']  # Target for Seats Won

# Step 2: Preprocess Features (X) - Categorical and Numerical
categorical_features = ['PM Candidate', 'Alliance_Part of NDA', 'Alliance_Regional focus']
numerical_features = [col for col in X.columns if col not in categorical_features]

# Preprocessing steps: OneHotEncode categorical features and StandardScale numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Step 3: Train-Test Split
X_train, X_test, y1_train, y1_test = train_test_split(X, y1, test_size=0.2, random_state=42)
X_train, X_test, y2_train, y2_test = train_test_split(X, y2, test_size=0.2, random_state=42)

# Step 4: Create Pipelines for Model 1 (Vote Share Prediction) and Model 2 (Seats Won Prediction)
model1_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', XGBRegressor(random_state=42))  # You can replace with RandomForestRegressor or LinearRegression
])

model2_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(random_state=42))  # You can replace with XGBRegressor or LinearRegression
])

# Step 5: Train Models
model1_pipeline.fit(X_train, y1_train)
model2_pipeline.fit(X_train, y2_train)

# Step 6: Predict and Evaluate Models
y1_pred = model1_pipeline.predict(X_test)
y2_pred = model2_pipeline.predict(X_test)

# Model 1 Evaluation (Vote Share)
print("Model 1 (Vote Share) Evaluation:")
print(f"R^2 Score: {r2_score(y1_test, y1_pred)}")
print(f"RMSE: {mean_squared_error(y1_test, y1_pred, squared=False)}")

# Model 2 Evaluation (Seats Won)
print("\nModel 2 (Seats Won) Evaluation:")
print(f"R^2 Score: {r2_score(y2_test, y2_pred)}")
print(f"RMSE: {mean_squared_error(y2_test, y2_pred, squared=False)}")
```
